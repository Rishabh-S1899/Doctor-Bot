Reading all the json data files using pandas

```python
import pandas as pd
import json
```

```python
def readAsDataframe(filePath):
    with open(filePath,"r") as f:
        dataFile = json.load(f)
    # print(len(dataFile))
    # return pd.DataFrame(dataFile)
    cnt = 0
    for item in dataFile:
        if cnt == 100:
            break
        cnt+=1
        print(item,dataFile[item])
```

```python
readAsDataframe("./CCF_data_convo.json")
```

```python
def write_numbers_to_file(input_filename, output_filename):
    try:
        # Open the input file in read mode
        with open(input_filename, 'r') as file_input:
            # Read the content of the file
            numbers = [int(line.strip()) for line in file_input]

            print(len(numbers))

    except FileNotFoundError:
        print("The input file does not exist")
    except Exception as e:
        print("An error occurred: ", str(e))


# Example usage
input_filename = './diabetes_indices.txt'
output_filename = 'numbers.txt'

write_numbers_to_file(input_filename, output_filename)
```
